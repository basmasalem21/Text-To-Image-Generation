```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python

model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"
```

```python
pipe1 = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe1 = pipe1.to("cuda")
```

```python
pipe2 = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe2 = pipe2.to("cuda")
```

```python
prompt= """lion in jungle
"""
```

```python
image1 = pipe1(prompt).images[0]

image2 = pipe2(prompt).images[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(image1)
plt.axis('off')
```

```python
plt.imshow(image2)
plt.axis('off')
```
